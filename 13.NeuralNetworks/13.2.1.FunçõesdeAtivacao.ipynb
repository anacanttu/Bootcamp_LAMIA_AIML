# Step Function

```python
def step_function(soma):
  if soma >= 1:
    return 1
  return 0
```

```python
step_function(-10)
```

```python
step_function(10)
```

```python
step_function(1)
```

# Sigmoid Function

```python
def sigmoid_function(soma):
  return 1/(1+ np.exp(-soma))
```

```python
sigmoid_function(-10)
```

```python
sigmoid_function(10)
```

```python
sigmoid_function(1)
```

# Hyperbolic Tangent Function

```python
def tahn_function(soma):
  return (np.exp(soma) - np.exp(-soma)) / (np.exp(soma)+ np.exp(-soma))
```

```python
tahn_function(-10)
```

```python
tahn_function(10)
```

```python
tahn_function(1)
```

# RELU Function

```python
def relu_function(soma):
  if soma >=0:
    return soma
  return 0
```

```python
relu_function(-20)
```

```python
relu_function(0)
```

```python
relu_function(10)
```

```python
relu_function(100)
```

# Softmax Function

```python
def softmax_function(x):
    ex = np.exp(x)
    return ex / ex.sum()
```

```python
softmax_function([7.0, 2.0, 1.3])
```

```python
softmax_function([2.0, 2.0, 5.3])
```
